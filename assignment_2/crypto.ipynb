```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import time
```

```python
cmc = requests.get('https://coinmarketcap.com')
soup = BeautifulSoup(cmc.content, 'html.parser')
# used to see structure of website 
print(soup.prettify())
```

```python
# trick to scrape historical data that free API does not provide 

data = soup.find('script', id="__NEXT_DATA__",type="application/json")
coins = {}

coin_data = json.loads(data.contents[0])
listings = coin_data['props']['initialState']['cryptocurrency']['listingLatest']['data']

for i in listings:
    coins[str(i['id'])] = i['slug']
```

```python
for i in coins:
    (time.sleep(3))
    page = requests.get(f'https://coinmarkercap.com/currencies/{coins[i]}/historical-data/?start=20200101&end=20201214')
    soup = BeautifulSoup(page.content, 'html.parser')
    data = soup.find('script', id="__NEXT_DATA__", type="application/json")
    if data != None:
        historical_data = json.loads(data.contents[0])
    quotes = historical_data['props']['initialState']['cryptocurrency']['ohlcvHistorical'][i]['quotes']
    info = historical_data['props']['initialState']['cryptocurrency']['ohlcvHistorical'][i]
    
       
```

```python
market_cap = []
volume = []
timestamp = []
name = []
symbol = []
slug = []
```

```python
for j in quotes:
    market_cap.append(j['quote']['USD']['market_cup'])
    volume.append(j['quote']['USD']['volume'])
    timestamp.append(j['quote']['USD']['timestamp'])
    name.append(info['name'])
    symbol.append(info['symbol'])
    slug.append(coins[i])
```

```python
df = pd.DataFrame(columns = ['marketcap', 'voulme', 'timestamp','name','symbol', 'slug'])
df['marketcap'] = market_cap
df['volume'] = volume
df['timestamp'] = timestamp
df['name'] = name
df['symbol'] = symbol
df['slug'] = slug 
df.to_cvs('criptoes', index = False)
```
